```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 400. # High res
```

```python
df_clean = pd.read_csv('Data/df_clean_3.csv')
df_clean['EDUCATION_CATEGORY'] = 'none'
df_clean[['EDUCATION', 'EDUCATION_CATEGORY']].head()
```

```python
education_mapping = {1: 'graduate school',
                    2: 'university',
                    3: 'high school',
                    4: 'others'}
```

```python
df_clean['EDUCATION_CATEGORY'] = df_clean['EDUCATION'].map(education_mapping)
df_clean[['EDUCATION', 'EDUCATION_CATEGORY']].head(10)
```

```python
edu_one_hot = pd.get_dummies(df_clean['EDUCATION_CATEGORY'])
edu_one_hot.head(10)
```

```python
df_with_category = pd.concat([df_clean, edu_one_hot], axis=1)
df_with_category[['EDUCATION_CATEGORY', 'graduate school', 'high school', 'university', 'others']].head(10)
```

```python
df_with_category.to_csv('Data/cleaned_data.csv')
```
